```python
from osgeo import gdal as gd
import matplotlib.pyplot as plt  
import numpy as np 
```

```python
data_set = gd.Open(r'imageries/guwahati/guw.tif')  
#print(data_set.RasterCount)  

band_1 = data_set.GetRasterBand(1) # NIR
band_2 = data_set.GetRasterBand(2) # Red  
band_3 = data_set.GetRasterBand(3) # Green
b1 = band_1.ReadAsArray()  
b2 = band_2.ReadAsArray()  
b3 = band_3.ReadAsArray()  
#print(b1)  
```

```python
fig = plt.figure()
ax = fig.add_subplot( projection='3d')
ax.scatter(b3, b2, b1, c=['g'],s=1)
ax.set_zlabel('B5_NIR')
ax.set_ylabel('B4_Red')
ax.set_xlabel('B3_Green')
ax.set_title('Scatter plot all bands')
plt.show()
```

```python
def scatplot(b1,b2,st):
    plt.scatter(b1,b2,s=1)
    plt.xlabel(st[1])
    plt.ylabel(st[2])
    plt.title(st[0])
scatplot(b1,b2,['Band 5 vs Band 4','Band 5','Band 4'])
```

```python
scatplot(b2,b3,['Band 4 vs Band 3','Band 4','Band 3'])
```

```python
scatplot(b3,b1,['Band 3 vs Band 5','Band 3','Band 5'])
```
